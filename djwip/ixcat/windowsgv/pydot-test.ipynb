```python
import pydot
from IPython.display import Image
```

```python
g = pydot.Dot()
g.add_node( pydot.Node( 'node!' ) )
g.add_edge( pydot.Edge('A','B') )
g.add_edge( pydot.Edge('A','C') )
g.add_edge( pydot.Edge( ('D','E') ) )
g.add_edge( pydot.Edge( ('E','F') ) )
g.add_edge( pydot.Edge( ('C','F') ) )
Image(g.create_png())
```

```python
import networkx as nx
to_pydot = nx.drawing.nx_pydot.to_pydot
```

```python
G = nx.Graph()
G.add_edge('networkx', 'pydotplus', weight=4)
G.add_edge('pydotplus', 'image', weight=2)
```
